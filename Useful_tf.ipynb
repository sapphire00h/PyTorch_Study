```python
from torchvision import transforms
import numpy as np
from PIL import Image
from torch.utils.tensorboard import SummaryWriter 

import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'

img_path="./imgs/mountain.jpg"
img=Image.open(img_path)
print(img)
```

## 使用`totensor`

```python
writer=SummaryWriter("logs")
```

```python
trans_totensor=transforms.ToTensor()
img_tensor=trans_totensor(img)
```

```python
writer.add_image("mountain",img_tensor)
writer.close()
```

## `Normalize`的使用

```python
#归一化，图片是RGB三个信号通道，需要提供三个标准差
#output[channel] = (input[channel] - mean[channel]) / std[channel]

writer=SummaryWriter("logs")
print(img_tensor[0][0][0])
trans_norm=transforms.Normalize([0.5,0.5,0.5],[0.5,0.5,0.5])
img_norm=trans_norm(img_tensor)
print(img_norm[0][0][0])
writer.add_image("mountain_norm",img_tensor)
writer.close()
```

## Resize

```python
#Resize
print(img.size)
trans_resize=transforms.Resize((512,512))
#img PIL->resize->img_resize PIL
img_resize=trans_resize(img)
print(img_resize.size)
```

## Compose - resize

```python
writer=SummaryWriter("logs")
```

```python
#等比缩放
trans_resize_2=transforms.Resize(1024)
#PIL->PIL->tensor
#compose相当于步骤的合成，后面函数的输入类型应该与前面函数的输出类型一致
#此处Compose相当于将图片进行等比缩放后再将图片转换为Trnsor类型
trans_compose=transforms.Compose([trans_resize_2,trans_totensor])
img_resize_2=trans_compose(img)
img_resize_2
```

```python
writer.add_image("mountain",img_resize_2,2)
writer.close()
```
